```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vit_s = [97.35, 92.475, 20.2375]
resnet_s = [96.1, 91.4625, 38.55]
x = ["Test Dataset", "Grayscale Dataset", "Cue Conflicted Dataset"]
plt.plot(x, vit_s)
plt.plot(x, resnet_s)
plt.title("Model Semantic Biases")
plt.ylabel("Test Accuracy")
plt.legend(["ViT", "ResNet"])
plt.show()
```

```python
resnet_l = [96.1, 91.225, 42.8125, 14.475]
vit_l = [97.35, 96.575, 85.9625, 9.1625]
x = ["Test Dataset", "Translated Dataset", "Masked Dataset", "Occluded Dataset"]
plt.plot(x, vit_l)
plt.plot(x, resnet_l)
plt.title("Model Locality Biases")
plt.ylabel("Test Accuracy")
plt.legend(["ViT", "ResNet"])
plt.show()
```

```python
vit_dg = [21.45584118096208, 75.48828125, 93.7125748502994, 52.98634812286689]
resnet_dg = [48.71468567065411, 60.15625, 91.01796407185628, 50.46928327645051]
labels = ["sketch", "art", "photo", "cartoon"]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, vit_dg, width, label="ViT")
plt.bar(x + width/2, resnet_dg, width, label="ResNet")

plt.xticks(x, labels)
plt.legend()
plt.title("Domain Generalisation with different Test Sets")
plt.xlabel("Test Domain")
plt.ylabel("Test Accuracy")
plt.show()
```
